# **Materi Pengulangan (Looping) dalam Python**

Dalam pemrograman, pengulangan (looping) digunakan untuk menjalankan kode secara berulang hingga kondisi tertentu tercapai. Python memiliki dua jenis loop utama:

1. FOR Loop

2. WHILE Loop


## Pengulangan dengan FOR

***for*** digunakan ketika kita tahu berapa kali perulangan akan dilakukan, atau ketika kita ingin melakukan iterasi pada koleksi data seperti list, string, atau range angka.

Bentuk Umum:

```
for variabel in range(start, stop, step):
    pernyataan
```
Keterangan:
start â†’ nilai awal (opsional, default 0)

stop â†’ batas akhir (tidak termasuk)

step â†’ loncatan (opsional, default 1)



````python
# Membuat urutan Angka 0 - 10

for i in range(0, 10, 2):
    print(i)
````

````python
# menggunakan fungsi Input

awal = int(input("Masukkan angka awal: "))
akhir = int(input("Masukkan angka akhir: "))
step = int(input("Masukkan step: "))

for i in range(awal, akhir + 1, step):
    print(i)
````

````python
# Mengoutputkan angka 1-6

for i in range(1, 6):
    print(i)
````

````python
# menggunakan fungsi Input

jumlah = int(input("Mau mengulang berapa kali? "))

for i in range(jumlah):
    print("Perulangan ke-", i+1)
````

````python
# Loop pada List

list_buah = ["apel", "mangga", "jeruk"]

for buah in list_buah:
    print("Saya suka",buah)
````

Penjelasan Code diatas:

Variabel buah adalah sebuah list berisi tiga elemen: "apel", "mangga", "jeruk".

Python akan menyimpan daftar item ini dalam urutan yang sama dengan nama "list_buah"

```
for buah in list_buah:
```
Baris ini artinya:

"Untuk setiap elemen di dalam list_buah, simpan elemen tersebut ke dalam variabel sementara bernama item."

Loop akan berjalan sebanyak jumlah elemen dalam list, yaitu 3 kali.

Urutan iterasinya:

item = "apel"

item = "mangga"

item = "jeruk"

```
print("Saya suka", buah)
```
Pada setiap perulangan, Python menjalankan perintah print().

Teks "Saya suka" akan digabungkan dengan nilai item saat itu.

````python
list_buah = []

jumlah = int(input("Berapa banyak buah? "))

for i in range(jumlah):
    buah = input(f"Masukkan nama buah ke-{i+1}: ")
    list_buah.append(buah)

for b in list_buah:
    print("Saya suka", b)
````

# **Pengulangan dengan WHILE**

"The for loop takes a collection of items and executes a block of code once
for each item in the collection. In contrast, the while loop runs as long as,
or while, a certain condition is true."

while digunakan ketika kita belum tahu jumlah perulangannya, tetapi berdasarkan kondisi tertentu.

Bentuk Umum

```
while kondisi:
    pernyataan
```
Loop akan terus berjalan selama kondisi bernilai True.


````python
i = 1
while i <= 5:
    print(i)
    i += 1
````

````python
# Program Input Berulang

teks = ""

while teks != "stop":
    teks = input("Ketik sesuatu (ketik 'stop' untuk berhenti): ")
````

````python
# While sebagai Counter

count = 10

while count > 0:
    print("Hitung mundur:", count)
    count -= 1   #sama dengan (count = count - 1)
````

````python
#infinit looping
count = 10

while count > 0:
    print("Hitung mundur:", count)
````

````python
# Using break to Exit a Loop

for i in range(1, 10):
    if i == 5:
        print("Loop dihentikan saat i =", i)
        break
    print("i =", i)
````

````python
total = 0

for i in range(1, 10):
    total += i
    if total > 10:
        print("Total sudah lebih dari 10. Loop dihentikan.")
        break
    print("Total sementara:", total)
````

````python
buah = ["apel", "mangga", "jeruk", "pisang"]

for b in buah:
    if b == "jeruk":
        print("Jeruk ditemukan, loop dihentikan.")
        break
    print("Memeriksa:", b)
````
