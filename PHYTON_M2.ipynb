# **List dalam Python**

Apa itu List?

List adalah struktur data di Python yang dapat menyimpan banyak nilai dalam satu variabel. Elemen dalam list bisa berupa angka, teks, atau bahkan campuran keduanya.



````python
bicycles = ['trek', 'cannondale', 'redline', 'specialized']

print(bicycles)
````

````python
print(bicycles[0])

# Index Positions Start at 0, Not 1
````

````python
#By asking for the item at index -1, Python always returns the last item in the list:

print(bicycles[-1])
````

# Changing, Adding, and Removing Elements

***Modifying Elements in a List***

````python
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)
````

````python
motorcycles[0] = 'ducati'
print(motorcycles)
````

***Adding Elements to a List***

````python
#Appending Elements to the End of a List

motorcycles = ['honda', 'yamaha', 'suzuki']

motorcycles.append('ducati')
print(motorcycles)
````

````python
# ANOTHER EXAMPLE FOR MAKE A LIST USING APPEND

motorcycles = []
motorcycles.append('honda')
motorcycles.append('yamaha')
motorcycles.append('suzuki')
print(motorcycles)
````

````python
# Inserting Elements into a List

motorcycles.insert(0, 'ducati')
print(motorcycles)
````

***Removing Elements from a List***

````python
# Removing an Item Using the del Statement

del motorcycles[0]
print(motorcycles)
````

````python
#  Removing an Item Using the pop() Method

motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)

popped_motorcycle = motorcycles.pop() #DEFAULT REMOVING THE LAST ITEM BUT U CAN USE INDEX TO REMOVE IT
print(motorcycles)
print(popped_motorcycle)
````

````python
# Removing an Item by Value

motorcycles = ['honda', 'yamaha', 'suzuki', 'ducati']
print(motorcycles)

motorcycles.remove('ducati')
print(motorcycles)
````

# Organizing a List

````python
# Sorting a List Permanently with the sort() Method

cars = ['bmw', 'audi', 'toyota', 'subaru']
cars.sort()
print(cars)
````

````python
cars.sort(reverse=True)
print(cars)
````

````python
#  Printing a List in Reverse Order
cars = ['bmw', 'audi', 'toyota', 'subaru']
print(cars)

cars.reverse()
print(cars)
````

````python
#  Finding the Length of a List
cars = ['bmw', 'audi', 'toyota', 'subaru']
len(cars)
````

````python
# Slicing a List

players = ['charles', 'martina', 'michael', 'florence', 'eli']
print(players[0:3]) # [(n-1):(n-1)]
````

# **Pengkondisian (IF–ELSE) dalam Python**

**Pengertian**

Pengkondisian digunakan untuk membuat keputusan dalam program.
Program akan mengeksekusi bagian kode tertentu hanya jika suatu kondisi terpenuhi (True).

Bentuk Dasar:
```
if kondisi:
    #Kode jika kondisi benar

else:
    #Kode jika kondisi salah
```




    


````python
#Simple if Statements

age = 19
if age >= 18:
  print("You are old enough to vote!")
````

````python
 #if-else Statements

age = int(input("Input Your Age: "))
if age >= 18:
    print("You are old enough to vote!")
    print("Have you registered to vote yet?")
else:
    print("Sorry, you are too young to vote.")
    print("Please register to vote as soon as you turn 18!")
````

# The if-elif-else Chain
Many real-world situations involve more than two possible conditions.
For example, consider an amusement park that charges different rates for
different age groups:

 •	Admission for anyone under age 4 is free.

 •	Admission for anyone between the ages of 4 and 18 is $25.

 •	Admission for anyone age 18 or older is $40.

 How can we use an if statement to determine a person’s admission rate?

````python
age = int(input("Input Your Age: "))

if age < 4:
    print("Your admission cost is $0.")

elif age < 18:
    print("Your admission cost is $25.")

else:
    print("Your admission cost is $40.")
````

````python
# Program konversi suhu dari Celcius ke Fahrenheit

#input
celcius = float(input("Masukkan suhu dalam Celcius: "))

#proses
fahrenheit = (9/5) * celcius + 32

#output
print("Suhu saat ini adalah", fahrenheit, "derajat Fahrenheit")
````

Katakanlah kita ingin **memodifikasi program** tersebut untuk mencetak peringatan saat cuaca ekstrem.

**Suhu di atas 90 derajat Fahrenheit** dan **di bawah 60 derajat Fahrenheit** akan menyebabkan peringatan cuaca panas dan dingin.

````python

#input
celcius = float(input("Masukkan suhu dalam Celcius: "))

#proses
fahrenheit = (9/5) * celcius + 32
if fahrenheit > 90:
    print("Awas Suhu Sedang Panas! Perbanyak Minum Air Putih")
elif fahrenheit < 60:
    print("Awas Suhu sedang Dingin! Pakailah pakaian yang Tebal")
else:
    print("Suhu Normal!")
````

#  Using if Statements with Lists

````python
 magicians = ['alice', 'david', 'carolina']

for magician in magicians:
  #Kedua Print ini masih didalam Loop, maka dia akan terus Mengulang
    print(f"{magician.title()}, that was a great trick!")
    print(f"I can't wait to see your next trick, {magician.title()}.\n")

#print ini ada diluar loop, maka hanya keluar jika looping sudah selesai
print("Thank you, everyone. That was a great magic show!")
````

````python
# Using Multiple Lists

available_toppings = ['mushrooms', 'olives', 'green peppers',
                      'pepperoni', 'pineapple', 'extra cheese']

requested_toppings = ['mushrooms', 'french fries', 'extra cheese']

for requested_topping in requested_toppings:
     if requested_topping in available_toppings:
       print(f"Adding {requested_topping}.")
     else:
        print(f"Sorry, we don't have {requested_topping}.")

print("\nFinished making your pizza!")
````
