**2.Variabel dalam Python**

Apa itu Variabel?

Variabel adalah wadah untuk menyimpan nilai/data yang dapat digunakan kembali di dalam program.

Aturan Penamaan Variabel:
1. Tidak boleh diawali dengan angka
2. Tidak boleh mengandung spasi (gunakan _ sebagai pengganti)
3. Bersifat case sensitive (huruf besar dan kecil dianggap berbeda)

```python
nama = "Syifa"
umur = 25
tinggi = 160.5
is_student = True

print(nama)
print(umur)
print(tinggi)
print(is_student)
```

3. Simple Data Types (Tipe Data Sederhana)

Python memiliki beberapa tipe data dasar:
| Tipe Data | Contoh            | Keterangan         |
| --------- | ----------------- | ------------------ |
| `int`     | 10, -5            | Bilangan bulat     |
| `float`   | 3.14, -2.5        | Bilangan desimal   |
| `str`     | "Hello", 'Python' | Teks/String        |
| `bool`    | True, False       | Logika benar/salah |


```python
# Integer dan Float
a = 10
b = 3.5
hasil = a + b
print("Hasil penjumlahan:", hasil)

# String
kata = "Belajar Python"
print(kata.upper())  # Mengubah huruf menjadi kapital semua

# Boolean
x = True
y = False
print("Nilai x:", x)
print("Nilai y:", y)
```

**4. List dalam Python**

Apa itu List?

List adalah struktur data di Python yang dapat menyimpan banyak nilai dalam satu variabel.
Elemen dalam list bisa berupa angka, teks, atau bahkan campuran keduanya.

```python
## Cara Membuat List:

buah = ["apel", "jeruk", "mangga", "Anggur"]
angka = [1, 2, 3, 4, 5]
campuran = ["Syifa", 25, True, 3.14]
```

```python
##Mengakses Elemen List:

print(buah[0])  # Menampilkan elemen pertama (apel)
print(buah[2])  # Menampilkan elemen ketiga (mangga)
```

```python
## Mengubah, Menambah, dan Menghapus Elemen List

# Ubah elemen
buah[1] = "pisang"
print(buah)

# Tambah elemen di akhir
buah.append("melon")
print(buah)

# Hapus elemen tertentu
buah.remove("apel")
print(buah)
```

```python
## Looping (Perulangan) dalam List
# Menampilkan semua elemen dalam list

for i in buah:
    print(i)
```

```python
# Operasi Lain pada List

angka = [10, 20, 30, 40, 50]

print("Jumlah data:", len(angka))      # Menghitung banyaknya elemen
print("Nilai maksimum:", max(angka))   # Nilai terbesar
print("Nilai minimum:", min(angka))    # Nilai terkecil
print("Rata-rata:", sum(angka)/len(angka))  # Rata-rata
```

```python
Nama = ["Riza", "Rafi", "Reza", "Rista"]

for n in Nama:
    print(n)
```
