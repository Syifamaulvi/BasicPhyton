# **LINEAR AND BINARY SEARCH**

## 1. Linear Search

**Linear Search** adalah metode pencarian dengan cara:
- Membandingkan **target** dengan setiap elemen di dalam list, dari indeks awal hingga akhir.
- Tidak membutuhkan data terurut.
- Cocok untuk data yang kecil dan sederhana.

Contoh use-case:  
> Mengecek apakah sebuah **kode barang** ada pada daftar barang di gudang.


```python
suppliers = ["Alpha Logistics", "Nusantara Cargo", "Star Express",
             "Logisindo", "BlueLine", "Nusantara Cargo"]
```

```python
def linear_search_supplier(data, supplier):
    lokasi = []
    for i in range(len(data)):
        if data[i] == supplier:
            lokasi.append(i)
    return lokasi

hasil = linear_search_supplier(suppliers, "Nusantara Cargo")
print("Hasil pencarian:", hasil)
```

```python
# Data awal: kode barang (SKU) dan stok
kode_barang = [1010, 1004, 1025, 1001, 1015, 1030, 1008]
stok_barang = [50, 20, 15, 30, 10, 5, 40]

print("Daftar Kode Barang:", kode_barang)
print("Daftar Stok:", stok_barang)
```

```python
def linear_search(data, target):
    """
    Mengembalikan list indeks tempat target ditemukan.
    Jika tidak ditemukan, mengembalikan list kosong.
    """
    posisi = []
    for i in range(len(data)):
        if data[i] == target:
            posisi.append(i)
    return posisi
```

```python
# Contoh pemanggilan
# print("Daftar Kode Barang:", kode_barang)
cari = int(input("Masukkan kode barang yang ingin dicari: "))

hasil = linear_search(kode_barang, cari)

if hasil:
    print("Kode barang ditemukan pada indeks:", hasil)
else:
    print("Kode barang TIDAK ditemukan.")
```

## 2. Binary Search

**Binary Search** hanya dapat digunakan jika data **sudah terurut** (sorted).

Cara kerja:
1. Ambil indeks tengah (mid).
2. Bandingkan nilai tengah dengan target:
   - Jika sama → ketemu.
   - Jika target lebih kecil → cari di bagian kiri.
   - Jika target lebih besar → cari di bagian kanan.
3. Diulangi sampai ketemu atau range habis.

Keunggulan:
- Jauh lebih cepat dibanding Linear Search untuk data besar.
- Kompleksitas waktu: `O(log n)`.

⚠️ **Syarat wajib:** data harus dalam **urutan naik** (ascending) atau turun (descending) konsisten.


```python
lead_time = [1, 2, 3, 4, 5, 7, 10, 12]
```

```python
def binary_search(data, target):
    low = 0
    high = len(data) - 1
    while low <= high:
        mid = (low + high) // 2
        if data[mid] == target:
            return mid
        elif target < data[mid]:
            high = mid - 1
        else:
            low = mid + 1
    return -1

print(binary_search(lead_time, 7))
```

```python
# Kita akan buat salinan dan urutkan kode_barang
kode_barang_sorted = sorted(kode_barang)
print("Data kode_barang (sorted):", kode_barang_sorted)
```

```python
def binary_search(data, target):
    low = 0
    high = len(data) - 1

    while low <= high:
        mid = (low + high) // 2
        if data[mid] == target:
            return mid  # mengembalikan indeks

        elif target < data[mid]:
            high = mid - 1
        else:
            low = mid + 1

    return -1  # tidak ditemukan
```

```python
# Contoh penggunaan:
# print("Data terurut:", kode_barang_sorted)
cari = int(input("Masukkan kode barang (untuk Binary Search): "))

pos = binary_search(kode_barang_sorted, cari)

if pos != -1:
    print("Kode barang ditemukan pada indeks (di data sorted):", pos)
else:
    print("Kode barang TIDAK ditemukan (Binary Search).")
```
