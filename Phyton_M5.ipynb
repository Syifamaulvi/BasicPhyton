# **Materi: Dictionary dalam Python**

A dictionary in Python is a collection of key-value pairs. Each key is connected to a value, and you can use a key to access the value associated with that key. A key’s value can be a number, a string, a list, or even another dictionary. In fact, you can use any object that you can create in Python as a value in a dictionary.

Dictionary adalah koleksi pasangan key-value (kunci–nilai) di Python.

Ciri-ciri Dictionary:

1. Berisi pasangan key : value
2. Menggunakan tanda kurung kurawal { }
3. Key harus unik
4. Value boleh duplikat
5. Bisa menampung berbagai tipe data

```python
mahasiswa = {
    "nama": "Syifa",
    "nim": "20250123",
    "jurusan": "Teknik Industri"
}

print(mahasiswa)
```

# Membuat Dictionary

```python
# Menggunakan { }
produk = {
    "kode": "A001",
    "nama": "Bolpoin",
    "harga": 5000
}

print(produk)
```

```python
buku = dict(judul="Python Dasar", halaman=250)

print(buku)
```

# Mengakses Nilai dalam Dictionary

```python
# Menggunakan key

print(produk["nama"])
```

```python
# Menggunakan get()

print(produk.get("harga"))
```

# Menambah & Mengubah Data

```python
# Menambah key-value

produk["stok"] = 120

print(produk)
```

```python
# Mengubah nilai

produk["harga"] = 6000

print(produk)
```

```python
#Menghapus Data

produk.pop("stok")       # hapus berdasarkan key
del produk["harga"]      # cara lain
produk.clear()           # hapus semua data
```

# Looping Through a Dictionary

```python
for key in produk:
    print(key)
```

```python
for value in produk.values():
    print(value)
```

```python
for key, value in produk.items():
    print(key, ":", value)
```

# Nested Dictionary (Dictionary di dalam Dictionary)

```python
gudang = {
    "A001": {"nama": "Bolpoin", "stok": 120},
    "A002": {"nama": "Pensil", "stok": 200}
}

print(gudang["A001"]["stok"])
```

# Fungsi-Fungsi Penting pada Dictionary

| Fungsi     | Kegunaan                       |
| ---------- | ------------------------------ |
| `len()`    | Menghitung jumlah item         |
| `keys()`   | Menampilkan semua key          |
| `values()` | Menampilkan semua value        |
| `items()`  | Menampilkan pasangan key-value |
| `update()` | Menggabungkan dictionary       |


```python
barang = {
    "kode": "B001",
    "nama": "Kopi Hitam",
    "harga": 12000,
    "stok": 50
}

print("Nama Barang:", barang["nama"])
barang["stok"] += 10
print("Stok Baru :", barang["stok"])
```
